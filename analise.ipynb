```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Caminho do arquivo
file_path = r"/Users/valentinaspohr/Desktop/tudo/unicamp/2s:2024/rouxinol/dados/Results/res 3GHz cap 1fF dB.txt"

# Leitura e extração dos dados
with open(file_path, "r") as file:
    lines = file.readlines()

# Extrair dados com regex
data = []
pattern = re.compile(r"^\s*([\d\.]+)\s+([-]?\d+\.\d+)")
for line in lines:
    match = pattern.match(line)
    if match:
        freq = float(match.group(1))
        db = float(match.group(2))
        data.append((freq, db))

# Criar DataFrame
df = pd.DataFrame(data, columns=["Frequency (GHz)", "S21 (dB)"])
# Encontrar ponto de máximo
max_point = df.loc[df["S21 (dB)"].idxmax()]

# Calcular largura de banda (3 dB abaixo do máximo)
half_max = max_point["S21 (dB)"] - 3
df_below = df[df["S21 (dB)"] >= half_max]
f_low = df_below["Frequency (GHz)"].min()
f_high = df_below["Frequency (GHz)"].max()
bandwidth = f_high - f_low

# Plotagem com valores na legenda
plt.figure(figsize=(10, 6))
plt.plot(df["Frequency (GHz)"], df["S21 (dB)"], label='S21 (dB)')
plt.axvline(x=max_point["Frequency (GHz)"], color='red', linestyle='--',
            label=f'Pico: {max_point["Frequency (GHz)"]:.7f} GHz')
plt.axhline(y=half_max, color='green', linestyle='--', label='-3 dB')
plt.axvline(x=f_low, color='purple', linestyle='--', label=f'F. inferior: {f_low:.7f} GHz')
plt.axvline(x=f_high, color='purple', linestyle='--', label=f'F. superior: {f_high:.7f} GHz')
plt.xlabel("Frequency (GHz)")
plt.ylabel("S21 (dB)")
plt.title("Ressonador de 3 GHz com capacitor de 1 fF")
plt.grid(True)
plt.legend(title=f'Largura de banda: {bandwidth:.3f} GHz')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Caminho para o arquivo de magnitude
file_path_mag = r"/Users/valentinaspohr/Desktop/tudo/unicamp/2s:2024/rouxinol/dados/Results/res 3GHz cap 1fF mag.txt"

# Ler o arquivo
with open(file_path_mag, "r") as file:
    lines = file.readlines()

# Extrair os dados
data_mag = []
pattern = re.compile(r"^\s*([\d\.]+)\s+([\d\.Ee+-]+)")
for line in lines:
    match = pattern.match(line)
    if match:
        freq = float(match.group(1))
        mag = float(match.group(2))
        data_mag.append((freq, mag))

# Criar DataFrame
df_mag = pd.DataFrame(data_mag, columns=["Frequency (GHz)", "S21 Magnitude"])

# Encontrar ponto de máximo
max_point_mag = df_mag.loc[df_mag["S21 Magnitude"].idxmax()]

# Calcular largura de banda a 70.7% do valor máximo (equivalente a -3 dB)
threshold_mag = max_point_mag["S21 Magnitude"] * 0.707
df_bw_mag = df_mag[df_mag["S21 Magnitude"] >= threshold_mag]
f_low_mag = df_bw_mag["Frequency (GHz)"].min()
f_high_mag = df_bw_mag["Frequency (GHz)"].max()
bandwidth_mag = f_high_mag - f_low_mag

# Plotagem
plt.figure(figsize=(10, 6))
plt.plot(df_mag["Frequency (GHz)"], df_mag["S21 Magnitude"], label='|S21|')
plt.axvline(x=max_point_mag["Frequency (GHz)"], color='red', linestyle='--',
            label=f'Pico: {max_point_mag["Frequency (GHz)"]:.6f} GHz')
plt.axhline(y=threshold_mag, color='green', linestyle='--', label='70.7% do pico')
plt.axvline(x=f_low_mag, color='purple', linestyle='--', label=f'F. inf: {f_low_mag:.6f} GHz')
plt.axvline(x=f_high_mag, color='purple', linestyle='--', label=f'F. sup: {f_high_mag:.6f} GHz')
plt.xlabel("Frequency (GHz)")
plt.ylabel("|S21|")
plt.title("Ressonador de 3 GHz com capacitor de 1 fF — Módulo de S21")
plt.grid(True)
plt.legend(title=f'Largura de banda: {bandwidth_mag:.6f} GHz')
plt.tight_layout()
plt.show()

```
