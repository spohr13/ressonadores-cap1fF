```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Caminho do arquivo
file_path = r"/Users/valentinaspohr/Desktop/tudo/unicamp/2s:2024/rouxinol/dados/Results/res 3GHz cap 1fF dB.txt"

# Leitura e extração dos dados
with open(file_path, "r") as file:
    lines = file.readlines()

# Extrair dados com regex
data = []
pattern = re.compile(r"^\s*([\d\.]+)\s+([-]?\d+\.\d+)")
for line in lines:
    match = pattern.match(line)
    if match:
        freq = float(match.group(1))
        db = float(match.group(2))
        data.append((freq, db))

# Criar DataFrame
df = pd.DataFrame(data, columns=["Frequency (GHz)", "S21 (dB)"])

# Plotar
plt.figure(figsize=(10, 6))
plt.plot(df["Frequency (GHz)"], df["S21 (dB)"])
plt.xlabel("Frequency (GHz)")
plt.ylabel("S21 (dB)")
plt.title("Ressonador de 3 GHz com capacitor de 1 fF")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Caminho para o arquivo de magnitude
file_path_mag = r"/Users/valentinaspohr/Desktop/tudo/unicamp/2s:2024/rouxinol/dados/Results/res 3GHz cap 1fF mag.txt"

# Ler o arquivo
with open(file_path_mag, "r") as file:
    lines = file.readlines()

# Extrair os dados
data_mag = []
pattern = re.compile(r"^\s*([\d\.]+)\s+([\d\.Ee+-]+)")
for line in lines:
    match = pattern.match(line)
    if match:
        freq = float(match.group(1))
        mag = float(match.group(2))
        data_mag.append((freq, mag))

# Criar DataFrame
df_mag = pd.DataFrame(data_mag, columns=["Frequency (GHz)", "S21 Magnitude"])

# Plotar
plt.figure(figsize=(10, 6))
plt.plot(df_mag["Frequency (GHz)"], df_mag["S21 Magnitude"])
plt.xlabel("Frequency (GHz)")
plt.ylabel("|S21|")
plt.title("Ressonador de 3 GHz com capacitor de 1 fF — Módulo de S21")
plt.grid(True)
plt.tight_layout()
plt.show()
```
